```python
import pandas as pd
import numpy as np

train = pd.read_csv("../input/PM25_train.csv") #load the dataset
test = pd.read_csv("../input/PM25_test.csv")
submission = pd.read_csv('../input/submission.csv')
train['Date'] = pd.to_datetime(train['Date'])
```

## 取得經緯度

```python
for i in range(0,252):
    index=test.index[test['device_id'] == submission.iloc[i]['device_id']][0]
    lat=test.iloc[index]['lat']
    lon=test.iloc[index]['lon']
    print("["+str(lon)+", "+str(lat)+"],")
```

## 取得每個測站平均

```python
for i in range(0,252):
    mean=train[train['device_id'] == submission.iloc[i]['device_id']]['PM2.5'].mean()
    print(str(mean)+", ")
```

```python
# 取得前10天平均
for i in range(0,252):
    data=train[train['device_id'] == submission.iloc[i]['device_id']]
    max=int(data.shape[0]*0.23)
    tot=0
    for j in range(0,max):
        tot+=data.iloc[j]['PM2.5']
    tot/=max
    print(str(tot)+", ")
#print(np.percentile(train[train['device_id'] == submission.iloc[0]['device_id']]['PM2.5'],1))
```
